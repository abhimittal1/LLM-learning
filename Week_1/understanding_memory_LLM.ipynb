```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
model = "gpt-4.1-mini"  
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key)   

messages = [
    {"role": "system", "content": "You are a helpful assistant."},  
    {"role": "user", "content": "hello i am a student studying LLMs."}
]

response = client.chat.completions.create(
    model=model,
    messages=messages
)

print(response.choices[0].message.content)
```

```python

# Asking questions regarding the previous context
messages = [
    {"role": "system", "content": "You are a helpful assistant."},  
    {"role": "user", "content": "Who Am i?"}
]

response = client.chat.completions.create(
    model=model,
    messages=messages
)

print(response.choices[0].message.content)
```

```python

# Asking questions regarding the previous context but now giving the context again
messages = [
    {"role": "system", "content": "You are a helpful assistant."},  
    {"role": "user", "content": "hello i am a student studying LLMs."},
    {"role": "assistant", "content": "Hello! That’s great to hear that you’re studying large language models (LLMs). How can I assist you with your studies today? Do you have any specific questions or topics you’d like to discuss?"},
    {"role": "user", "content": "Who Am i?"}
]

response = client.chat.completions.create(
    model=model,
    messages=messages
)

print(response.choices[0].message.content)
```
